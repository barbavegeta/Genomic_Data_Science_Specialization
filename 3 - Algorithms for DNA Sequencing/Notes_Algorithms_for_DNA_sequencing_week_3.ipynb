```python
def hammingDistance(x, y):
    nmm = 0
    for i in xrange(0, len(x)):
        if x[i] != y[i]:
            nmm += 1
    return nmm
```

```python
def edDistRecursive(a,b):
    if len(a) == 0:
        return len(b)
    if len(b) == 0:
        return len(a)
    delt = 1 if a[-1] != b[-1] else 0
    return min(edDistRecursive(a[ :- 1], b[ :- 1]) + delt,
               edDistRecursive(a, b[ :- 1]) + 1,
               edDistRecursive(a[ :- 1], b) + 1)
```

```python
def editDistance(x, y):
    # Create distance matrix
    D = []
    for i in range(len(x) + 1):
        D.append([0] * (len(y) + 1))  # Fixed matrix initialization
    # Initialize first row and column of matrix
    for i in range(len(x) + 1):
        D[i][0] = i
    for i in range(len(y) + 1):  # Corrected loop variable
        D[0][i] = i
    # Fill in the rest of the matrix    
    for i in range(1, len(x) + 1):
        for j in range(1, len(y) + 1):
            distHor = D[i][j - 1] + 1
            distVer = D[i - 1][j] + 1  # Corrected variable name
            if x[i - 1] == y[j - 1]:
                distDiag = D[i - 1][j - 1]
            else:
                distDiag = D[i - 1][j - 1] + 1  # Fixed assignment
            D[i][j] = min(distHor, distVer, distDiag)
    # Edit distance is the value in the bottom right corner of the matrix    
    return D[-1][-1]
```

```python
alphabet = ['A', 'C', 'G', 'T']
score = [[0,4,2,4,8], \
        [4,0,4,2,8], \
        [2,4,0,4,8],\
        [8,8,8,8,8]]
```

```python
def globalAlignment(x, y):
    D = []
    for i in range(len(x) + 1):
        D.append([0] * (len(y) + 1))  # Fixed matrix initialization
    
    for i in range(1, len(x) + 1):
        D[i][0] = D[i-1][0] + score[alphabet.index(x[i-1])][-1]
    for i in range(1, len(y) + 1):  # Corrected loop variable
        D[0][i] = D[0][i-1] + score[-1][alphabet.index(y[i-1])]
    
    for i in range(1, len(x) + 1):
        for j in range(1, len(y) + 1):
            distHor = D[i][j - 1] + score[-1][alphabet.index(y[j-1])]
            distVer = D[i - 1][j] + score[alphabet.index(x[i-1])][-1]
            if x[i - 1] == y[j - 1]:
                distDiag = D[i - 1][j - 1]
            else:
                distDiag = D[i - 1][j - 1] + score[alphabet.index(x[i-1])][alphabet.index(y[j-1])]
            
            D[i][j] = min(distHor, distVer, distDiag)
    
    return D[-1][-1]
```

```python
def overlap(a, b, min_length=3):
    start = 0
    
    while True:
        start = a.find(b[:min_length], start)
        if start == -1:
            return 0
        if b.startswith(a[start:]):
            return len(a)-start
        start += 1
```

```python
def naive_overlap_map(reads, k):
    olaps = {}
    for a,b in permutations(reads,2):
        olen = overlap(a,b,min_length=k)
        if olen >0:
            olaps[(a,b)] = olen
        return olaps
```

```python
reads = ['ACGGATGATC', 'GATCAAGT', 'TTCACGGA']
print(naive_overlap_map(reads,3))
```
